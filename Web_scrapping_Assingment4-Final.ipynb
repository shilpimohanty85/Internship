1.Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re

import warnings
warnings.filterwarnings("ignore")
```

```python
#connecting to the web driver
driver=webdriver.Chrome(r'C:\Users\91845\chromedriver.exe') 
time.sleep(2)

# Opening the wikipedia.com
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(2)

# Make empty lists
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]

 #scraping rank of most-viewed YouTube videos
try:
    Rank= driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[1]")
    for i in Rank[:30]:
        rank.append(i.text)
        time.sleep(1)
except NoSuchElementException:
    print("Exception Occured.No such element")
    
#scraping name
try:
    Name= driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[2]")
    for i in Name[:30]:
        name.append(i.text)
        time.sleep(1)
except NoSuchElementException:
    print("Exception Occured. No such element")
        
#scraping artist 
try:
    Artist= driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[3]")
    for i in Artist[:30]:
        artist.append(i.text)   
        time.sleep(1) 
except NoSuchElementException:
    print("Exception Occured. No such element")        
    
#scraping upload_date    
try:
    Date=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[4]")
    for i in Date[:30]:
        upload_date.append(i.text)
        time.sleep(1)  
except NoSuchElementException:
    print("Exception Occured. No such element")   
    
#scraping views
try:
    Views= driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[5]")
    for i in Views[:30]:
        views.append(i.text)    
        time.sleep(1) 
except NoSuchElementException:
    print("Exception Occured. No such element")
    
#creating dataframe
df=pd.DataFrame({'Rank':rank,
                'Name':name,
                'Artist':artist,
                'Upload_date':upload_date,
                'Views':views})
df    
    
```
