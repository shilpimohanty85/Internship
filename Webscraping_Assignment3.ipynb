```python
# getting all the urls to open each page from our above search results
urls =[]
for i in range(0,3): # upto 3 pages
    for x in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
        urls.append(x.get_attribute('href'))
    driver.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']").click()
    time.sleep(3)
    
# all the initial empty lists
brand=[]
prod_name=[]
rating=[]
rating_num=[]
price= []
return_ex=[]
exp_del = []
availability=[]
other_det = []

# obtaining all the details one after another
for i in urls[:150]:
    driver.get(i)
    time.sleep(3)
    
    # scrapping the brand name
    try:
        brand_name=driver.find_element_by_xpath("//td[@class='a-span9']/span")
        brand.append(brand_name.text)
    except NoSuchElementException:
        brand.append('-')    
                
      # scrapping the product name
    try:
        product=driver.find_element_by_id("productTitle")
        prod_name.append(product.text)
    except NoSuchElementException:
        prod_name.append('-')         
        
      # scrapping the rating
    try:
        ratestar=driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']//span")
        rating.append(ratestar.text)
    except NoSuchElementException:
        rating.append("No Rating")       
                
      # scrapping the rating number
    try:
        ratenum=driver.find_element_by_id("acrCustomerReviewText")
        rating_num.append(ratenum.text)
    except NoSuchElementException:
        rating_num.append('-')         
                
      # scrapping the price
    try:
        prod_price=driver.find_element_by_xpath('//td[@class="a-span12"]')
        price.append(prod_price.text)
    except NoSuchElementException:
        price.append('-')      
                
     # scrapping the return/exchange
    try:
        ret_exch=driver.find_element_by_xpath("/html/body/div[1]/div[2]/div[10]/div[6]/div[1]/div[4]/div/div/div/div/div/form/div/div/div/div/div[13]/div/div/span/span/a")
        return_ex.append(ret_exch.text)#get_attribute("href"))
    except NoSuchElementException:
        return_ex.append('-') 
        
     # scrapping the expected delivery
    try:
        ex_de=driver.find_element_by_xpath("//span[@class='a-text-bold']")
        exp_del.append(ex_de.text)
    except NoSuchElementException:
        exp_del.append('-')           
               
     # scrapping the availability
    try:
        availa=driver.find_element_by_id("availability")
        availability.append(availa.text)
    except NoSuchElementException:
        availability.append('-')       
                           
           
     # getting the other details
    try:
        othr_det=driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']/li/span")
        other_det.append(othr_det.text)
    except NoSuchElementException:
        other_det.append('-')
        
# creating the entire dataframe now
df = pd.DataFrame()
df['Brand Name'] = brand
df['Product Name'] = prod_name
df['Rating'] = rating
df['Rating Number'] = rating_num
df['Price'] = price
df['Return/Exchange'] = return_ex
df['Expected Delivery'] = exp_del
df['Availability'] = availability
df['Product URLs'] = urls
df['Other Details'] = other_det     
df        
```

```python
hostel_name=[]
try:
    name_div=driver.find_elements_by_xpath("//h2[@class='title title-6']") 

    for i in name_div:
         hostel_name.append(i.text) 
except NoSuchElementException:
        distance.append('-')    
hostel_name[:10]
```

```python
distance=[]
try:
    distance_div=driver.find_elements_by_xpath("//span[@class='description']") 

    for i in distance_div:
        distance.append(i.text) 
except NoSuchElementException:
        distance.append('-') 
distance[:10]        
   
```

```python
rating=[]
try:
    rating_div=driver.find_elements_by_xpath("//div[@class='score orange big']") 
    for i in rating_div:
        rating.append(i.text) 
except NoSuchElementException:
    distance.append('-')
rating[:10]    
   
```

```python
review=[]
try:
    review_div=driver.find_elements_by_xpath("//div[@class='reviews']") 
    for i in review_div:
        review.append(i.text) 
except NoSuchElementException:
    distance.append('-')
review[:10]
```

```python
overall=[]
try:
    overall_div=driver.find_elements_by_xpath("//div[@class='keyword']") 
    for i in overall_div:
        overall.append(i.text) 
except NoSuchElementException:
        description.append('-')
        
overall[:10]
```

```python
pvt_price=[]
try:
    pvt_div=driver.find_elements_by_xpath("//a[@class='prices-col']//div[1]//div") 
    for i in pvt_div:
        pvt_price.append(i.text) 
except NoSuchElementException:
       description.append('-')
            
    
pvt_price[:10]
```

```python
dorm_price=[]
dorm_div=driver.find_elements_by_xpath("//a[@class='prices-col']//div[2]//div") 

for i in dorm_div:
    dorm_price.append(i.text) 
    time.sleep(1)
dorm_price[:10]
```

```python
facilities=[]
try:
    fac1_div=driver.find_elements_by_xpath("//span[@class='label']") 
    fac2_div=driver.find_elements_by_xpath("//div[@class='has-sanitation']") 

    for i in fac1_div:
        for j in fac2_div:
            facilities.append(i.text +','+ j.text) 
except NoSuchElementException:
     facilities.append('-')
```

```python
description=[]
desc_div=driver.find_elements_by_xpath("//div[@class='rating-factors prop-card-tablet rating-factors small']") 

for i in desc_div:
    description.append(i.text) 
    time.sleep(1)
description[:10]
```
